```python
from transformers import GPT2Tokenizer, GPT2Model,AutoModel,AutoTokenizer
import torch
model_name="dnagpt/gene_eng_gpt2_v1_ft"
device="cuda"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model.to(device)

def get_text_embedding(text):
    """
    使用 GPT-2 模型获取文本的向量表示。
    
    参数:
        text (str): 输入文本。
        model_name (str): 预训练 GPT-2 模型名称，默认为 "gpt2"。
        device (str): 设备名称（"cpu" 或 "cuda"）。
    
    返回:
        torch.Tensor: 文本的向量表示，维度为 [hidden_size]。
    """

    # 将文本编码为输入 ID 并添加批量维度
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=256)
    input_ids = inputs["input_ids"].to(device)
    attention_mask = inputs["attention_mask"].to(device)
    
    # 获取模型的隐藏层输出
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
        hidden_states = outputs.last_hidden_state  # [batch_size, seq_length, hidden_size]
    
    # 平均池化：获取序列中所有词向量的平均值
    embeddings = hidden_states.mean(dim=1).squeeze()  # [hidden_size]
    
    return embeddings
```

```python
para_list = []
for line in open("NC_000001.11_chapter_1.fna.p1"):
    line = line.strip()
    line = "".join(line.split())
    para_list.append(line)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def dynamic_clustering_sequential(
    paragraphs, paragraph_vect_list, similarity_threshold=0.5, max_cluster_size=50
):
    """
    对段落进行动态聚类，仅判断是否归入当前类别，形成顺序的段落分组。

    参数:
        paragraphs (list): 输入的段落列表。
        paragraph_vect_list (list): 输入段落的嵌入向量列表。
        similarity_threshold (float): 相似度阈值。
        max_cluster_size (int): 每个类别的最大段落数量。

    返回:
        list: 聚类后的段落组（每组是段落的索引列表）。
        list: 每个类别的中心向量。
    """
    clusters = []  # 保存类别，每个类别是段落索引的列表
    cluster_centers = []  # 保存每个类别的中心向量

    current_cluster = []
    current_center = None

    for idx, vector in enumerate(paragraph_vect_list):
        # 将 CUDA Tensor 转为 CPU 并转换为 NumPy 数组
        if isinstance(vector, torch.Tensor):
            vector = vector.cpu().numpy()

        if 0== len(current_cluster):
            # 初始化第一个类别
            current_cluster.append(idx)
            current_center = vector
        else:
            # 计算当前段落与当前类别中心的相似度
            similarity = cosine_similarity([vector], [current_center])[0][0]

            if similarity >= similarity_threshold and len(current_cluster) < max_cluster_size:
                # 加入当前类别
                current_cluster.append(idx)
                # 更新类别中心向量
                current_center = np.mean(
                    [
                        paragraph_vect_list[i].cpu().numpy()
                        if isinstance(paragraph_vect_list[i], torch.Tensor)
                        else paragraph_vect_list[i]
                        for i in current_cluster
                    ],
                    axis=0,
                )
            else:
                # 结束当前类别并开始新类别
                print("current idx", idx, "create new cluster len", len(current_cluster))
                clusters.append(current_cluster)
                cluster_centers.append(current_center)
                current_cluster = [idx]
                current_center = vector

    # 将最后一个类别加入结果
    if current_cluster:
        clusters.append(current_cluster)
        cluster_centers.append(current_center)

    # 将聚类结果中的索引转换为段落
    grouped_paragraphs = [[paragraphs[i] for i in cluster] for cluster in clusters]
    return grouped_paragraphs, cluster_centers
```

```python
grouped_paragraphs, cluster_centers = dynamic_clustering_sequential(para_list, paragraph_vect_list)

print("聚类后段落数量", len(cluster_centers))
# 查看结果
# for i, group in enumerate(grouped_paragraphs):
#     print(f"Group {i + 1}:")
#     print(" ".join(group[:3]) + "...")  # 打印每组前 3 个段落
```

```python
def combine_clusters_and_centers(clusters, cluster_centers):
    """
    将聚类结果和中心点合并成结构化数据。

    参数:
        clusters (list): 聚类后的段落组，每组是一个段落列表。
        cluster_centers (list): 每个类别的中心向量。

    返回:
        list: 每个类别包含中心向量和段落列表的字典。
    """
    combined_data = []
    for center, para_list in zip(cluster_centers, clusters):
        combined_data.append({
            "cluster_center": center.tolist(),  # 将 NumPy 数组转为列表
            "para_list": para_list
        })
    return combined_data

# 使用示例
combined_data = combine_clusters_and_centers(grouped_paragraphs, cluster_centers)

# 检查输出
print(combined_data[11])  # 查看第一个类别的内容
```
